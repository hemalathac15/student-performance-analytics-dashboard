```python
#Report card analysis bar chart
import matplotlib.pyplot as plt
import numpy as np
#dummy data of class 7
students=["Ram","Hari","Shiva","Divya","Ramya","Keshav"]
Maths= [82,87,91,94,96,99]
Science= [88,78,80,85,90,87]
English= [80,79,82,86,81,78]
Social= [82,80,78,83,84,81]
Sanskrit= [80,78,76,75,82,79]
subjects= ["Maths","Science","English","Social","Sanskrit"]
Total_subject_marks=np.array([Maths,Science,English,Social,Sanskrit])
x=np.arange(len(students))
bar_width=0.15
plt.figure(figsize=(12, 6))
#Group plotting
n=len(subjects)
offset_base= (n-1) /2
#Loop for each subject to plot the bar with apt offset
for i, subject in enumerate(subjects):
    factor=i-offset_base
    offset=factor*bar_width
    plt.bar(x+offset,Total_subject_marks[i],width=bar_width,label=subject)
plt.xticks(x,students)
plt.xlabel("Students",fontsize=12)
plt.ylabel("Marks",fontsize=12)
plt.title("Marks Of Each Students In Different Subjects",fontsize=12,weight="bold")
plt.suptitle("REPORT CARD ANALYTICS",fontsize=12,weight="bold")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.minorticks_on()
plt.grid(which="major",linestyle="-",linewidth=0.5,color="r")
plt.grid(which="minor",linestyle=":",linewidth=0.5,color="k")
plt.show()
```

```python
#Report Card Analysis Line graph
import matplotlib.pyplot as plt
import numpy as np
#dummy data of class 7
students=["Ram","Hari","Shiva","Divya","Ramya","Keshav"]
Maths= [82,87,91,94,96,99]
Science= [88,78,80,85,90,87]
English= [80,79,82,86,81,78]
Social= [82,80,78,83,84,81]
Sanskrit= [80,78,76,75,82,79]
subjects= ["Maths","Science","English","Social","Sanskrit"]
Total_subject_marks=np.array([Maths,Science,English,Social,Sanskrit])
plt.figure(figsize=(12, 7))
#Loop for each subject to plot as line
for i, subject in enumerate(subjects):
    plt.plot(students, Total_subject_marks[i], marker='o', linestyle='-', label=subject)
plt.xlabel("Students",fontsize=12)
plt.ylabel("Marks",fontsize=12)
plt.title("Marks Of Each Students In Different Subjects",fontsize=12,weight="bold")
plt.suptitle("REPORT CARD ANALYTICS",fontsize=12,weight="bold")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.minorticks_on()
plt.grid(True, which="major",linestyle="-",linewidth=0.5,color="r")
plt.grid(True,which="minor",linestyle=":",linewidth=0.5,color="k")
plt.tight_layout()
plt.show()
```

```python
#Data Distribution of Class Marks in Maths
import matplotlib.pyplot as plt
import numpy as np
Maths= [82, 87, 91, 94, 96, 99] 
plt.figure(figsize=(10,5))
bins=np.arange(82,100,5)
#defining mark ranges eg: 90 <= marks > 85 (1 student)
plt.hist( Maths,
          bins=[82, 87, 91, 94, 96, 99],
          edgecolor='k',
          color='b',
          rwidth=0.9
        )
plt.xlabel("Maths Marks")
plt.ylabel("Number of students (Frequency)")
plt.suptitle("REPORT CARD ANALYTICS",fontsize=10,weight="bold")
plt.title("Distribution of Class Marks in Maths",fontsize=10,weight="bold")
plt.xticks(bins)
plt.grid(True,axis='y', linestyle='--', alpha=0.7,color="r")
plt.show()
```

```python
#Scatter plot of Maths Vs Science
import matplotlib.pyplot as plt
import numpy as np
x= np.array([82, 87, 91, 94, 96, 99])
y= np.array([88,78,80,85,90,87])
plt.figure(figsize=(10,5))
colors=np.array(['r','y','c','m','g','b'])
plt.scatter(x,y,c=colors,edgecolor="k")
plt.suptitle("REPORT CARD ANALYTICS",fontsize=10,weight="bold")
plt.title("Scatter Plot of Maths Vs Science",fontsize=10,weight="bold")
plt.xlabel("Maths")
plt.ylabel("Science")
plt.minorticks_on()
plt.grid(True, which="major",linestyle="-",linewidth=0.5,color="r")
plt.grid(True,which="minor",linestyle=":",linewidth=0.5,color="k")
plt.show()
```
